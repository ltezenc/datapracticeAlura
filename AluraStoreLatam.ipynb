### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



# 2. Ventas por categoría

```python
precio_list=tienda['Precio']
ingresa_total_tienda=0
for precio in precio_list:
  ingresa_total_tienda += int(precio)
print(f"ingresa_total de tienda {ingresa_total_tienda}")

precio_list1=tienda2['Precio']
ingresa_total_tienda2=0
for precio1 in precio_list1:
  ingresa_total_tienda2 += int(precio1)
print(f"ingreso total tienda2 {ingresa_total_tienda2}")

precio_list1=tienda3['Precio']
ingresa_total_tienda3=0
for precio1 in precio_list1:
  ingresa_total_tienda3 += int(precio1)
print(f"ingreso total tienda3 {ingresa_total_tienda3}")

precio_list1=tienda4['Precio']
ingresa_total_tienda4=0
for precio1 in precio_list1:
  ingresa_total_tienda4 += int(precio1)
print(f"ingreso total tienda4 {ingresa_total_tienda4}")
```

```python
list_categoria=tienda['Categoría del Producto']
conteo_categorias = list_categoria.value_counts()
top_categorias = conteo_categorias.head(3)
print(f"y las mas vendidas de la tienda son \n {top_categorias}")

list_categoria2=tienda2['Categoría del Producto']
conteo_categorias2 = list_categoria2.value_counts()
top_categorias2 = conteo_categorias2.head(3)
print(f"y las mas vendidas de la tienda2 son \n {top_categorias2}")

list_categoria3=tienda3['Categoría del Producto']
conteo_categorias3 = list_categoria3.value_counts()
top_categorias3 = conteo_categorias3.head(3)
print(f"y las mas vendidas de la tienda3 son \n {top_categorias3}")

list_categoria4=tienda4['Categoría del Producto']
conteo_categorias4 = list_categoria4.value_counts()
top_categorias4 = conteo_categorias4.head(3)
print(f"y las mas vendidas de la tienda4 son \n {top_categorias4}")
```

# 3. Calificación promedio de la tienda


```python
calificacion_list=tienda['Calificación']
calificacion_tienda=0
for nota in calificacion_list:
  calificacion_tienda += int(nota)
  valoracion=calificacion_tienda/len(calificacion_list)
print(f"la valoracion de la tienda {round(valoracion,2)}")

calificacion_list=tienda2['Calificación']
calificacion_tienda2=0
for nota in calificacion_list:
  calificacion_tienda2 += int(nota)
  valoracion2=calificacion_tienda2/len(calificacion_list)
print(f"la valoracion de la tienda2 {round(valoracion2,2)}")

calificacion_list=tienda3['Calificación']
calificacion_tienda3=0
for nota in calificacion_list:
  calificacion_tienda3 += int(nota)
  valoracion3=calificacion_tienda3/len(calificacion_list)
print(f"la valoracion de la tienda3 {round(valoracion3,2)}")

calificacion_list=tienda4['Calificación']
calificacion_tienda4=0
for nota in calificacion_list:
  calificacion_tienda4 += int(nota)
  valoracion4=calificacion_tienda4/len(calificacion_list)
print(f"la valoracion de la tienda4 {round(valoracion4,2)}")

```

# 4. Productos más y menos vendidos

```python
list_producto=tienda['Producto']
conteo_producto = list_producto.value_counts()
top_producto = conteo_producto.head(3)
print(f"y los productos mas vendidos de la tienda son \n {top_producto}")

list_producto2=tienda2['Producto']
conteo_producto2 = list_producto2.value_counts()
top_producto2 = conteo_producto2.head(3)
print(f"y los productos mas vendidos de la tienda2 son \n {top_producto2}")

list_producto3=tienda3['Producto']
conteo_producto3 = list_producto3.value_counts()
top_producto3 = conteo_producto3.head(3)
print(f"y los productos mas vendidos de la tienda3 son \n {top_producto3}")

list_producto4=tienda4['Producto']
conteo_producto4 = list_producto4.value_counts()
top_producto4 = conteo_producto4.head(3)
print(f"y los productos mas vendidos de la tienda4 son \n {top_producto4}")
```

```python
import pandas as pd

# Tienda 1
conteo_producto = tienda['Producto'].value_counts().head(3)
df_t1 = conteo_producto.reset_index()
df_t1.columns = ['Producto', 'Cantidad']
df_t1['Tienda'] = 'Tienda 1'

# Tienda 2
conteo_producto2 = tienda2['Producto'].value_counts().head(3)
df_t2 = conteo_producto2.reset_index()
df_t2.columns = ['Producto', 'Cantidad']
df_t2['Tienda'] = 'Tienda 2'

# Tienda 3
conteo_producto3 = tienda3['Producto'].value_counts().head(3)
df_t3 = conteo_producto3.reset_index()
df_t3.columns = ['Producto', 'Cantidad']
df_t3['Tienda'] = 'Tienda 3'

# Tienda 4
conteo_producto4 = tienda4['Producto'].value_counts().head(3)
df_t4 = conteo_producto4.reset_index()
df_t4.columns = ['Producto', 'Cantidad']
df_t4['Tienda'] = 'Tienda 4'

# Unir todos en un solo DataFrame
df_top_productos = pd.concat([df_t1, df_t2, df_t3, df_t4], ignore_index=True)

df_top_productos
```

# 5. Envío promedio por tienda

```python
envio_list=tienda['Costo de envío']
costo_total=0
for costo in envio_list:
  costo_total += int(costo)
  envio_promedio=costo_total/len(envio_list)
print(f"el costo de envio promedio de la tienda es  {round(envio_promedio)}")

envio_list2=tienda2['Costo de envío']
costo_total2=0
for costo2 in envio_list2:
  costo_total2 += int(costo2)
  envio_promedio2=costo_total2/len(envio_list2)
print(f"el costo de envio promedio de la tienda2 es  {round(envio_promedio2)}")

envio_list3=tienda3['Costo de envío']
costo_total3=0
for costo3 in envio_list3:
  costo_total3 += int(costo3)
  envio_promedio3=costo_total3/len(envio_list3)
print(f"el costo de envio promediode la tienda3 es  {round(envio_promedio3)}")

envio_list4=tienda4['Costo de envío']
costo_total4=0
for costo4 in envio_list4:
  costo_total4 += int(costo4)
  envio_promedio4=costo_total4/len(envio_list4)
print(f"el costo de envio promedio de la tienda4 es  {round(envio_promedio4)}")


```

```python
import matplotlib.pyplot as plt
import pandas as pd



tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingresa_total_tienda, ingresa_total_tienda2, ingresa_total_tienda3, ingresa_total_tienda4]
costo_envio = [envio_promedio, envio_promedio2, envio_promedio3, envio_promedio4]
calificacion = [valoracion, valoracion2, valoracion3, valoracion4]


# Crear DataFrame
df_resumen = pd.DataFrame({
    'Tienda': tiendas,
    'Ingresos': ingresos,
    'Costo_Envio_Prom': costo_envio,
    'Calificacion_Prom': calificacion
})

print(df_resumen)


plt.figure(figsize=(8, 6))
plt.bar(tiendas, calificacion, color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('calificacion')
plt.title('calificacion promedio de la tienda')
plt.grid(axis='y', linestyle='--')
plt.show()


plt.figure(figsize=(8, 6))
plt.bar(tiendas, ingresos, color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos Totales')
plt.title('Ingresos Totales por Tienda')
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(8, 6))
plt.bar(tiendas, costo_envio, color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envio')
plt.title('Costo envio Promedio por Tienda')
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(
    df_resumen['Ingresos'],
    df_resumen['Calificacion_Prom'],
    color='royalblue',
    s=100,
    alpha=0.7
)

# Etiquetas para que sea más informativo
for i, tienda in enumerate(df_resumen['Tienda']):
    plt.text(
        df_resumen['Ingresos'][i] + 0.02*df_resumen['Ingresos'][i],
        df_resumen['Calificacion_Prom'][i],
        tienda,
        fontsize=10
    )

plt.title('Ingresos vs. Calificación por Tienda')
plt.ylabel('Calificación Promedio')
plt.xlabel('Ingresos')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()


for t in df_top_productos['Tienda'].unique():
    df_temp = df_top_productos[df_top_productos['Tienda']==t]

    ax = df_temp.plot(
        kind='pie',
        y='Cantidad',
        autopct='%1.1f%%',
        title=f'Top productos más vendidos - {t}',
        figsize=(6,6),
        labels=['' for _ in df_temp['Producto']],  # Dejar etiquetas vacías en el gráfico
        legend=False
    )
    # Añadir leyenda externa
    ax.legend(
        df_temp['Producto'],
        title='Producto',
        loc='center left',
        bbox_to_anchor=(1.0, 0.5)
    )
    plt.ylabel('')  # Para que no aparezca "Cantidad"
    plt.tight_layout()
    plt.show()
```
